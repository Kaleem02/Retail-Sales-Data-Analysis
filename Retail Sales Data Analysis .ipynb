27. Create a new column Sales Category which categorizes Sales into 'Low', 
'Medium', and 'High' based on quantiles. (hint: use pd.qcut())

```python
# Create Sales Category column based on quantiles
df['Sales Category'] = pd.qcut(
    df['Quantity'], 
    q=3,  # 3 quantiles → Low, Medium, High
    labels=['Low', 'Medium', 'High']
)
print(df)
```

'Break Even', and 'Profit' based on the value of Profit.(use own values)

```python
df['Profit Category'] = np.where(df['Profit'] < 0, 'Loss',np.where(df['Profit'] == 0, 'Break Even', 'Profit'))
```

30. Create a new column Customer Segment which 
categorizes Segment into 'Individual', 'Corporate', and 'Home Office'. 
(use map()) 

This tells pandas how to translate values:

"Consumer" → "Individual"

"Corporate" → "Corporate"

"Home Office" → "Home Office"

🔎 Step 3: The .map() Function

.map() looks at each row in the Segment column.

It checks if that value exists as a key in segment_mapping.

If it does → replaces it with the value from the dictionary.

If it doesn’t → returns NaN (missing value).

```python
segment_mapping = {
    'Consumer': 'Individual',
    'Corporate': 'Corporate',
    'Home Office': 'Home Office'
}

# Create new column using map()
df['Customer Segment'] = df['Segment'].map(segment_mapping)
```
